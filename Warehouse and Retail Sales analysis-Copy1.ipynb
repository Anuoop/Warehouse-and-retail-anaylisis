```python
# Data manipulation
import pandas as pd
import numpy as np 

# Visualization
import seaborn as sns
import matplotlib.pyplot as plt

# Warning
import warnings
warnings.filterwarnings("ignore")
```

```python
data_raw = pd.read_csv('Warehouse_and_Retail_Sales.csv')
```

```python
data_raw.head()
```

```python
data_raw.describe()
```

```python
#Comment:
#Negative values in Retail Sales may be a hint for returned items
#75% percentil is at 3.267 and the maximum at 2739. We need to check for outliers
```

```python
# Handle missing values
data = data_raw.dropna()
data.reset_index(drop=True, inplace=True)

# Handle negative values
data = data[(data['RETAIL SALES'] >= 0) & (data['RETAIL TRANSFERS'] >= 0) & (data['WAREHOUSE SALES'] >= 0)]

data.describe()
```

```python
# Identify outlier
sns.boxplot(data= data, x='RETAIL SALES')
```

```python
# Filter Dataset - Outlier
Q1 = data['RETAIL SALES'].quantile(0.25)
Q3 = data['RETAIL SALES'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5* IQR
upper = Q3 + 1.5* IQR

outliers = data[(data['RETAIL SALES'] > upper) | (data['RETAIL SALES'] < lower)]
print(f"The dataset contains: {len(outliers)}.")

data[data['RETAIL SALES'].between(lower, upper)]
print(f"The filtered Dataset hase the shape: {data.shape}.")
```

```python
# Create date - column
data['Date'] = pd.to_datetime(data['YEAR'].astype(str) + '-' + data['MONTH'].astype(str), format='%Y-%m')
data.drop(columns = ['YEAR', 'MONTH'], inplace= True)
data.head()
```

```python
#Visualisation
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.histplot(data=data, x='RETAIL SALES', bins=500, kde=True)
plt.title('Distribution of Retail Sales')

plt.subplot(1, 2, 2)
sns.histplot(data=data, x='RETAIL SALES', bins=500, kde=True)
plt.xlim(0, 25)
plt.title('Zoom in Retail Sales')

plt.tight_layout()  
plt.show()
```

```python
plt.figure(figsize=(14, 6))
sns.lineplot(data=data, x="Date", y="RETAIL SALES")
plt.title('Sales per Month')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data = data, x='ITEM TYPE', order=data['ITEM TYPE'].value_counts().index, palette='Set2')
plt.title('Count of Items by Type')
plt.xlabel('Item Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```
